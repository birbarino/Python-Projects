# Data Types

Types are **quantitative** data (number of dogs seen) and **categorical** data (breed of dog)
- Categorical is used to group or label a set of item



Analysis techniques for quantitative - four main aspects to analyzing
- Measure of Center
- Measure of Spread
- Shape of data
- Outliers

Analyis techniques for Categorical data
- Looking at counts or proportion of items that fall into each group

Categorical data has two types
- Ordinal - "ranked" categories
  - Degree type, rating on survey, letter grades, etc.
- Nominal - no ranked order

Quantitative Data Types categories
- Discrete - number of pages in a book, number of dogs. countable values.
- Continuous - height, time, age, income. Anything that can be split and a smaller unit still exists.



# Measures of Center

Mean
- Not always the best due to strong influence from very small or very large values
- Always returns a decimal value
- AKA average/expected value.

Median
- Must sort values
- If even number of values, the median is the average of the two middle values

Using mean vs. median depends on the shape of dataset and presence of outliers

Mode
- Most common value in dataset
  - Might have multiple modes or no mode at all.

# Measures of Spread

Histograms - most common visual for quantitative data



- We can choose the bins as we want them, or let software figure it out.

5-Number Summary is used to calculate spread of data
- Max
- Min
- First Quartile (Q1)
- Second Quartile (Q2/median)
- Third Quartile (Q3)

The 5-Number Summary is great for when datasets are not symmetric.

When calculating...
1. Order the dataset
  - Min is first, max is last
  - Median is middle of dataset (Q2)
2. Q1 == median of data on left of Q2 (excluding Q2)
3. Q3 == median of data on right of Q2

*Note: if any of the datasets you are calculating the median are even in length/count, you just take the average of the two middle values. This will also mean that you will not need to average to get Q1 and Q3 as you will just use the values next to the middle, including those used to get Q2/median




Box plots especially use the 5-number summary



Range
- Measure the spread of the data. This is the max and min






Interquartile Range (IQR)
- Difference between Q3 and Q1 (75% percentile - 25% percentile)
- Get the quartiles with `pandas.DataFrame.quantile([.25, .5, .75])` or `pandas.Series.quantile(.25)`
- Use this (IQR) to find the Outlier Threshold (see below). Anything below or above the threshold would be considered an outlier

```
Outlier Threshold Equations
Upper bound = Q3 + 1.5 * IQR
Lower bound = Q1 - 1.5 * IQR
```

# Shape and Outliers

## Shape

- Right-skewed
  - Mean > median
  - Real world application: Time between phone calls at call center, time until light bulb dies, amount of drug remaining in bloodstream, distribution of wealth, human athletic abilities



- Left-skewed
  - Mean < median
  - Grades as a percentage in many universities, age of death, price changes.


  
- Symmetric (normal distribution, typically)
  - Mean == median
  - Height, weight, errors, precipitation

Distribution's mode is the tallest bar in the histogram

Box and whisker plot leans toward the skew (longer whisker on side of skew)



## Outliers

Points that fall very far from the rest of the data. It influences measures like mean and standard deviation much more than the 5-number summary measures.



Things to consider about outliers
- Noting they exist and the impact on summary statistics
- If typo - remove/fix
- Understand why they exist and how they impact the questions we are trying to answer
- 5-number summary is often better indication of spread/center than mean and S.Dev when outliers are present

When working with them:
- Plot data to ID outliers
- Handle accordingly via previous methods
- If no outliers and normal distribution is followed, use S.Dev and mean to describe and report that it's normally distributed.
- If you have a skew or outliers, use 5-number summary and report the outliers.
