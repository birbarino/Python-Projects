# Intro

## What is an Experiment

**Experiments** are comparisons between two+ groups, ensuring equivalence between all groups besides the one that is being manipulated
- The goal is to observe and record the effects of changing one feature/variable.

**Unit of analysis** is the thing under observation, like the user in a web experiment or something.

**Observational Studies** are defined by their lack of control. It's just taking observations and letting things happen naturally, no intervention from the recorder. AKA naturalistic or correlational studies. Because of the lack of control, we cannot infer causality, as unobserved factors may be at play. These studies do, however, help rule out alternative hypotheses as more and more data is collected.

**Quasi-experiments** are between observational studies and experiments. Some of the control requirements of a true experiment are met, e.g., testing some new supplemental materials in a school course. Choosing two different schools and giving one new materials does not also exert control over teaching methods, 1-1 interactions and teaching that occurs, student inter-teaching, and more. It also may not dictate exactly how the material is applied, what order it's used in, etc.

## Normal Distribution and Quadratics
Bell curves are often called a "normal distribution" and has two variables mean ($\mu$) and variance ($\sigma$)

For any outcome $x$, the quadratic difference between values $x$ and $\mu$ can be written as

$$(x - \mu)^2$$

Which looks like this:



When $x = \mu$, $f(x) = 0$ (as in, the result is 0 because 2-2=0)

The function is also strictly positive/never negative due to the squaring.

Additionally, the function is quadratic, so it will make a curve of some sort.

## Variance in Normal Distribution
Apply variance by dividing everything by $\sigma^2$

$$\frac{(x-\mu)^2}{\sigma^2}$$

The effect of variance occurs along the y-axis because it is the denominator

It effectively drags the top down, which will widen the graph

Wide variances yield wide quadratics (4, for example)

Tight variances yield sharp quadratics (.25, for example)



## Negative Sign in Normal Distribution
What is the shape of $- \frac{1}{2}\frac{(x-\mu)^2}{\sigma^2}$



Multiplying by -1/2 flips it along the Y axis (because the denominator is not 1) and the negative will invert all values

## Normalization of Normal Distribution
This is honestly the most succinct way I think this could be described



Basically, we normalize, or ensure that we can apply statistical methods and probability calculations, by throwing the $\sqrt{2\pi\sigma^2}$ under $1$ 

$$
1/\sqrt{2\pi\sigma^2}
$$

Then applying it to the function via multiplication. This is how the normal distribution can be written:

$$
N(x;u,\sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} * e^{-\frac{1}{2}\frac{(x-\u)^2}{\sigma^2}}
$$

The N here means "normalized". Mu is the mean and sigma squared is the variance of the normal distribution

Additionally, the quadratic term penalizes any deviations from the expectation $\mu$ and the exponential squeezes the shape back into the curves.

Shape is shown below. Y-axis is $p(x)$, or the probability of $x$. $x$ is the possible mean values of the experiments.
- Probability is maximized when $x = \mu$ (mean)
- Probability decreases as $x$ deviates from the mean
- The probability approaches 0 when $x$ is very far from the mean.

# Sampling Distributions and the Central Limit Theorem

## Descriptive vs. Inferential Statistics

### Descriptive Statistics
It's about describing the collected data using
- Measures of center
- Measures of spread
- Shape
- Distribution
- Outliers

Visuals of the data can also be used for better understanding 

### Inferential Statistics
Using collected data to draw conclusions to a larger population
- Requires taking a sample that accurately represents the population of interest
- Some terms we care about here
  - Population - entire group of interest
  - Parameter - numeric summary about a population
  - Sample - subset of popluation
  - Statistic - numeric summary about sample

### Sampling Distributions in Python

Using a seed to keep results the same

```python
np.random.seed(42)
v = np.random.randint(0, 12, size=10)

# print(v)
print("size:", v.size)
print("mean:", np.mean(v))
print("variance:", np.var(v))
print("std dev:", np.std(v).round(2))
```
